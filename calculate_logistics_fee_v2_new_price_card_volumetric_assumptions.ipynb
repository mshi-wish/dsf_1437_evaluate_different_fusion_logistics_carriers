```python
def calculate_price(w, channel, price_card):
    for k, v in price_card.items():
        if (channel == k[0]) and (w>k[1]) and (w<k[2]):
            return v['item_fee'] + w*v['kg_fee']
        if (channel == k[0]) and (w==k[1]):
            return v['item_fee'] + w*v['kg_fee']
    return 
```

```python
channels = {}

for c in ['1150', '1152', '2001', 'NewTD']:
    channels[c] = {}
    for w in np.arange(0, 30.01, 0.01):
        channels[c][round(w,2)] = calculate_price(w, c, price_card)
        
price_card_detail = pd.DataFrame(channels)

price_card_detail['1152_1150_ratio'] = price_card_detail['1152'] / price_card_detail['1150']
price_card_detail['NewTD_1150_ratio'] = price_card_detail['NewTD'] / price_card_detail['1150']
price_card_detail['NewTD_2001_ratio'] = price_card_detail['NewTD'] / price_card_detail['2001']
```

## good stores: monthly recombined packages (days) >= 3 and average weight > 1kg

```python
temp = data.groupby(['store_id', 'epc_combined_month']).agg({
            'epc_combined_date': 'count',\
            'weight': 'mean'
        }).reset_index().groupby('store_id').mean()

good_stores = set(temp[(temp['epc_combined_date']>=3) & (temp['weight']>=1)].index)

len(good_stores)
```

```python
def price_volumetric_weight_included(weight, price, volumetric_card, price5p5, price9):
    prob5p5 = volumetric_card[round(weight, 1)][5.5]['prob_of_pack']
    prob9 = volumetric_card[round(weight, 1)][9]['prob_of_pack']
    packed_price = max(price, prob5p5 * price5p5 + prob9 * price9 + (1 - prob5p5 - prob9) * price)
    
    return packed_price
    

def purchase_logistics_fee_lm_by_ratio(x, price_card_detail, volumetric_card, old_channel, new_channel, \
                                       stores=set(), q4_increase=1.2):
    ratio_col = f'{new_channel}_{old_channel}_ratio'
    ratio = price_card_detail.loc[round(x['weight'], 2)][ratio_col]
    price5p5 = price_card_detail.loc[5.50][f'{old_channel}'] * ratio # price of 5.5kg
    price9 = price_card_detail.loc[9.00][f'{old_channel}'] * ratio # price 9kg
    
    if x['order_channel_post'] == old_channel:
        if (len(stores)) == 0 or (len(stores) > 0 and x['store_id'] in stores): # use the new channel
            if not ratio: # no ratio of old channel / new channel
                price = x[f'purchase_logistics_fee_lm_{new_channel}']
            else:
                price = x['purchase_logistics_fee_lm_previous'] * ratio
                
            if x['weight'] >= 1.6: # consider price that the order is packed
                return price_volumetric_weight_included(\
                        x['weight'], price, volumetric_card, price5p5, price9)
            else: 
                return price
        
        else: # use the old channel
            return x['purchase_logistics_fee_lm_previous'] * q4_increase
    
    return





data['purchase_logistics_fee_lm_1150_1152_by_ratio'] = data.apply(\
            lambda x: purchase_logistics_fee_lm_by_ratio(\
            x, price_card_detail, volumetric_card, 1150, 1152, stores=good_stores, q4_increase=1.1),\
            axis=1)

data['purchase_logistics_fee_lm_1150_NewTD_by_ratio'] = data.apply(\
            lambda x: purchase_logistics_fee_lm_by_ratio(\
            x, price_card_detail, volumetric_card, 1150, 'NewTD', stores=good_stores, q4_increase=1.1),\
            axis=1)

data['purchase_logistics_fee_lm_2001_NewTD_by_ratio'] = data.apply(\
            lambda x: purchase_logistics_fee_lm_by_ratio(\
            x, price_card_detail, volumetric_card, 2001, 'NewTD', stores=good_stores, q4_increase=1.1),\
            axis=1)


# count how many packages are upgraded
data['is_good_store_id'] = data.apply(\
                lambda x: x['store_id'] if x['store_id'] in good_stores else 0, \
                axis=1)

data['fm_count_upgraded'] = data.apply(\
                lambda x: x['fm_count'] if x['store_id'] in good_stores else 0, \
                axis=1)

data['lm_count_upgraded'] = data.apply(\
                lambda x: x['lm_count'] if x['store_id'] in good_stores else 0, \
                axis=1)

data['num_mids_upgraded'] = data.apply(\
                lambda x: x['num_mids'] if x['store_id'] in good_stores else 0, \
                axis=1)

data['gmv_upgraded'] = data.apply(\
                lambda x: x['gmv'] if x['store_id'] in good_stores else 0, \
                axis=1)

data['weight_upgraded'] = data.apply(\
                lambda x: x['weight'] if x['store_id'] in good_stores else 0, \
                axis=1)
```

```python
monthly = data.groupby(['epc_combined_month', 'order_channel_post']).agg({
    'store_id': ['count', 'nunique'],\
    'fm_count': 'sum',\
    'lm_count': 'sum',\
    'num_mids': 'sum',\
    'gmv': 'sum',\
    'weight': 'sum',\
    
    'is_good_store_id': 'nunique',\
    'fm_count_upgraded': 'sum',\
    'lm_count_upgraded': 'sum',\
    'num_mids_upgraded': 'sum',\
    'gmv_upgraded': 'sum',\
    'weight_upgraded': 'sum',\
    
    'sales_fm': 'sum',\
    'purchase_fm': 'sum',\
    'purchase_logistics_fee_lm_previous': 'sum',\
    'purchase_other_fees_lm_previous': 'sum',\
    'logistics_profit_cny_previous': 'sum',\
    'purchase_logistics_fee_lm_1152': 'sum',\
    'purchase_logistics_fee_lm_newtd': 'sum',\
    'purchase_logistics_fee_lm_1150_1152_by_ratio': 'sum',\
    'purchase_logistics_fee_lm_1150_NewTD_by_ratio': 'sum',\
    'purchase_logistics_fee_lm_2001_NewTD_by_ratio': 'sum',\
})


monthly['logistics_profit_margin_previous'] = monthly['logistics_profit_cny_previous'] / monthly['sales_fm']


monthly['logistics_profit_margin_1150_1152_by_ratio'\
           ] = (monthly['sales_fm'] - \
                monthly['purchase_fm'] - \
                monthly['purchase_logistics_fee_lm_1150_1152_by_ratio'] - \
                5*monthly['purchase_other_fees_lm_previous']\
               ) / monthly['sales_fm']

monthly['logistics_profit_margin_1150_NewTD_by_ratio'\
           ] = (monthly['sales_fm'] - \
                monthly['purchase_fm'] - \
                monthly['purchase_logistics_fee_lm_1150_NewTD_by_ratio'] - \
                5*monthly['purchase_other_fees_lm_previous']\
               ) / monthly['sales_fm']

monthly['logistics_profit_margin_2001_NewTD_by_ratio'\
           ] = (monthly['sales_fm'] - \
                monthly['purchase_fm'] - \
                monthly['purchase_logistics_fee_lm_2001_NewTD_by_ratio'] - \
                5*monthly['purchase_other_fees_lm_previous']\
               ) / monthly['sales_fm']

monthly = monthly.reset_index()
monthly['logistics_profit_margin_1150_1152_by_ratio'] = monthly.apply(\
                                lambda x: np.nan if x[('order_channel_post', '')]!=1150 \
                                                 else x['logistics_profit_margin_1150_1152_by_ratio'],\
                                axis=1)

monthly['logistics_profit_margin_1150_NewTD_by_ratio'] = monthly.apply(\
                                lambda x: np.nan if x[('order_channel_post', '')]!=1150 \
                                                 else x['logistics_profit_margin_1150_NewTD_by_ratio'],\
                                axis=1)

monthly['logistics_profit_margin_2001_NewTD_by_ratio'] = monthly.apply(\
                                lambda x: np.nan if x[('order_channel_post', '')]!=2001 \
                                                 else x['logistics_profit_margin_2001_NewTD_by_ratio'].values[0],\
                                axis=1)


monthly[('is_good_store_id', 'nunique')] = monthly[('is_good_store_id', 'nunique')] / monthly[('store_id', 'nunique')]
monthly[('fm_count_upgraded', 'sum')] = monthly[('fm_count_upgraded', 'sum')] / monthly[('fm_count', 'sum')]
monthly[('lm_count_upgraded', 'sum')]  = monthly[('lm_count_upgraded', 'sum')] / monthly[('lm_count', 'sum')]
monthly[('num_mids_upgraded', 'sum')] = monthly[('num_mids_upgraded', 'sum')] / monthly[('num_mids', 'sum')]
monthly[('gmv_upgraded', 'sum')] = monthly[('gmv_upgraded', 'sum')] / monthly[('gmv', 'sum')]
monthly[('weight_upgraded', 'sum')] = monthly[('weight_upgraded', 'sum')] / monthly[('weight', 'sum')]
```
